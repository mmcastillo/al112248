```python
from scipy import signal
import matplotlib.pyplot as plt
#
#.              2
#  H(s) = --------------
#.        s^2 + 0.3s + 1

lti = signal.lti([2.0], [1,0.3, 1.0])

t, y = signal.step(lti)


plt.plot(t, y)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Respuesta al escalón 2o orden')
plt.grid()
```

```python
lti2 = signal.lti([2.0], [1,0.2, 1.0])
lti3 = signal.lti([2.0], [1,0.1, 1.0])
t, y2 = signal.step(lti2)
t, y3 = signal.step(lti3)
plt.plot(t, y, t, y2, t, y3)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Respuesta al escalón 2o orden')
plt.grid()
```
